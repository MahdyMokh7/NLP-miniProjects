```python
import re
from torch.utils.data import Dataset, DataLoader

def clean_and_tokenize(text):
    text = text.lower()
    
    text = re.sub(r'[^a-z0-9\s]', '', text)
    
    tokens = text.split()
    
    return tokens
```

```python
import pandas as pd
from collections import Counter

def build_vocab(texts, min_freq=2, unk_token="<UNK>"):
    if isinstance(texts, pd.Series):
        texts = texts.tolist() 
    
    all_tokens = []
    for tokens in texts:
        all_tokens.extend(tokens)

    word_counts = Counter(all_tokens)

    vocab_words = [word for word, count in word_counts.items() if count >= min_freq]

    vocab_words = [unk_token] + vocab_words

    word2idx = {word: idx for idx, word in enumerate(vocab_words)}
    idx2word = {idx: word for word, idx in word2idx.items()}

    return word2idx, idx2word
```

```python
def generate_cbow_pairs(tokens, window_size=2):
    """
    Generate CBOW pairs (context -> target) from a list of tokens.

    Args:
        tokens (list): List of tokens (strings or integers).
        window_size (int): Number of words to consider on each side of the target.

    Returns:
        List of tuples: (context_words, target_word)
    """
    pairs = []
    for i in range(len(tokens)):
        start = max(i - window_size, 0)
        end = min(i + window_size + 1, len(tokens))
        context = [tokens[j] for j in range(start, end) if j != i]
        target = tokens[i]
        pairs.append((context, target))
    return pairs


def generate_skipgram_pairs(tokens, window_size=2):
    """
    Generate Skip-gram pairs (target -> context) from a list of tokens.

    Args:
        tokens (list): List of tokens (strings or integers).
        window_size (int): Number of words to consider on each side of the target.

    Returns:
        List of tuples: (target_word, context_word)
    """
    pairs = []
    for i in range(len(tokens)):
        start = max(i - window_size, 0)
        end = min(i + window_size + 1, len(tokens))
        target = tokens[i]
        context_words = [tokens[j] for j in range(start, end) if j != i]
        for context in context_words:
            pairs.append((target, context))
    return pairs

```

```python
import torch
from torch.nn.utils.rnn import pad_sequence
import random

def collate_fn_cbow(batch, window_size, vocab_size, negative_samples):
    """
    Returns:
      context_tensor:  (N_pairs, ctx_len)
      target_tensor:   (N_pairs,)
      negative_tensor: (N_pairs, K)
    """
    context_list = []
    target_list = []
    negative_list = []

    for tokens in batch:
        pairs = generate_cbow_pairs(tokens, window_size)

        for context, target in pairs:
            context_list.append(torch.tensor(context, dtype=torch.long))

            target_list.append(int(target))

            # negative sampling
            negs = []
            for _ in range(negative_samples):
                neg = random.randint(1, vocab_size - 1)
                while neg == target:
                    neg = random.randint(1, vocab_size - 1)
                negs.append(neg)

            negative_list.append(torch.tensor(negs, dtype=torch.long))

    # pad contexts
    context_tensor = pad_sequence(context_list, batch_first=True, padding_value=0)
    target_tensor = torch.tensor(target_list, dtype=torch.long)
    negative_tensor = torch.stack(negative_list, dim=0)

    return context_tensor, target_tensor, negative_tensor


import torch
import random

def collate_fn_skipgram(batch, window_size, vocab_size, negative_samples):
    """
    Returns:
      target_tensor:   (N_pairs,)
      context_tensor:  (N_pairs,)
      negative_tensor: (N_pairs, K)
    """
    target_list = []
    context_list = []
    negative_list = []

    for tokens in batch:
        pairs = generate_skipgram_pairs(tokens, window_size)

        for target, context in pairs:
            target_list.append(int(target))
            context_list.append(int(context))

            # negative sampling
            negs = []
            for _ in range(negative_samples):
                neg = random.randint(1, vocab_size - 1)
                while neg == target:
                    neg = random.randint(1, vocab_size - 1)
                negs.append(neg)

            negative_list.append(torch.tensor(negs, dtype=torch.long))

    target_tensor = torch.tensor(target_list, dtype=torch.long)
    context_tensor = torch.tensor(context_list, dtype=torch.long)
    negative_tensor = torch.stack(negative_list, dim=0)

    return target_tensor, context_tensor, negative_tensor
```

```python
class WordEmbeddingDataset(Dataset):
    def __init__(self, data, word2idx):
        self.data = data
        self.word2idx = word2idx

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        tokens = self.data[idx]
        indices = [self.word2idx.get(w, self.word2idx['<UNK>']) for w in tokens]
        return indices
```

```python
import torch 
import torch.nn as nn 
import torch.optim as optim 
import torch.nn.functional as F 
from tqdm import tqdm 

class SkipgramNegativeSampling(nn.Module):
    def __init__(self, vocab_size, embedding_dim, negative_samples=5):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.context_layer = nn.Embedding(vocab_size, embedding_dim)  # Updated layer name

        self.vocab_size = vocab_size
        self.negative_samples = negative_samples

        # Xavier init gives better convergence
        nn.init.xavier_uniform_(self.embedding.weight)
        nn.init.xavier_uniform_(self.context_layer.weight)  # Consistent naming

    def forward(self, target_idx, context_idx, negative_idx):
        target_emb = self.embedding(target_idx)
        context_emb = self.context_layer(context_idx)  # Updated layer name

        # Calculate positive loss
        pos_score = torch.sum(target_emb * context_emb, dim=1)
        pos_loss = F.logsigmoid(pos_score)

        # NEGATIVE SAMPLES
        neg_emb = self.context_layer(negative_idx)  # Updated layer name
        neg_score = torch.bmm(neg_emb, target_emb.unsqueeze(2)).squeeze(2)
        neg_loss = torch.sum(F.logsigmoid(-neg_score), dim=1)

        loss = -(pos_loss + neg_loss).mean()
        return loss

    def train_model(self, train_loader, val_loader, epochs, lr=0.001, device="cpu"):
        self.to(device)
        optimizer = optim.Adam(self.parameters(), lr=lr)

        train_losses = []
        val_losses = []

        for epoch in range(epochs):
            print(f"\nEpoch [{epoch+1}/{epochs}]")

            # -----------------------------
            # TRAINING WITH PROGRESS BAR
            # -----------------------------
            self.train()
            total_train_loss = 0
            train_progress = tqdm(train_loader, desc="Training", leave=False)

            for target_idx, context_idx, negative_idx in train_progress:
                target_idx = target_idx.to(device)
                context_idx = context_idx.to(device)
                negative_idx = negative_idx.to(device)

                loss = self(target_idx, context_idx, negative_idx)

                optimizer.zero_grad()
                loss.backward()
                optimizer.step()

                total_train_loss += loss.item()
                train_progress.set_postfix(loss=loss.item())

            avg_train_loss = total_train_loss / len(train_loader)
            train_losses.append(avg_train_loss)

            # -----------------------------
            # VALIDATION WITH PROGRESS BAR
            # -----------------------------
            self.eval()
            total_val_loss = 0
            val_progress = tqdm(val_loader, desc="Validating", leave=False)

            with torch.no_grad():
                for target_idx, context_idx, negative_idx in val_progress:
                    target_idx = target_idx.to(device)
                    context_idx = context_idx.to(device)
                    negative_idx = negative_idx.to(device)

                    loss = self(target_idx, context_idx, negative_idx)
                    total_val_loss += loss.item()
                    val_progress.set_postfix(val_loss=loss.item())

            avg_val_loss = total_val_loss / len(val_loader)
            val_losses.append(avg_val_loss)

            print(f"Train={avg_train_loss:.4f} | Val={avg_val_loss:.4f}")

        return train_losses, val_losses

    def evaluate_test(self, test_dataloader, device="cpu"):
        self.eval()
        total_loss = 0

        with torch.no_grad():
            for target_idx, context_idx, negative_idx in test_dataloader:
                target_idx = target_idx.to(device)
                context_idx = context_idx.to(device)
                negative_idx = negative_idx.to(device)

                loss = self(target_idx, context_idx, negative_idx)
                total_loss += loss.item()

        return total_loss / len(test_dataloader)

    def get_word_embeddings(self):
        return self.embedding.weight.detach()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from tqdm import tqdm
import torch
import torch.optim as optim

class CBOWNegativeSampling(nn.Module):
    def __init__(self, vocab_size, embedding_dim, negative_samples=NEGATIVE_SAMPLES):
        super(CBOWNegativeSampling, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.context_layer = nn.Embedding(vocab_size, embedding_dim)
        self.negative_samples = negative_samples

    def forward(self, context_idx, target_idx, negative_idx):
        context_embeds = self.embedding(context_idx)
        context_embeds = context_embeds.mean(dim=1)
        target_embeds = self.context_layer(target_idx)
        negative_embeds = self.context_layer(negative_idx)
        
        positive_score = torch.sum(context_embeds * target_embeds, dim=1)
        negative_score = torch.bmm(negative_embeds, context_embeds.unsqueeze(2)).squeeze()
        negative_score = torch.sum(negative_score, dim=1)

        loss = torch.sum(F.logsigmoid(positive_score) + F.logsigmoid(-negative_score))
        return -loss

    def train_model(self, train_dataloader, val_dataloader, epochs, lr=0.001, device='cpu'):
        self.to(device)
        optimizer = optim.SGD(self.parameters(), lr=lr)
        train_losses = []
        val_losses = []

        for epoch in range(epochs):
            print(f"\nEpoch [{epoch+1}/{epochs}]")
            
            # -------------------------
            # TRAIN LOOP WITH tqdm
            # -------------------------
            self.train()
            total_train_loss = 0

            train_progress = tqdm(train_dataloader, desc="Training", leave=False)

            for context_idx, target_idx, negative_idx in train_progress:
                context_idx = context_idx.to(device)
                target_idx = target_idx.to(device)
                negative_idx = negative_idx.to(device)

                optimizer.zero_grad()
                loss = self(context_idx, target_idx, negative_idx)
                loss.backward()
                optimizer.step()

                total_train_loss += loss.item()
                train_progress.set_postfix(loss=loss.item())

            avg_train_loss = total_train_loss / len(train_dataloader)
            train_losses.append(avg_train_loss)

            # -------------------------
            # VALIDATION LOOP WITH tqdm
            # -------------------------
            self.eval()
            total_val_loss = 0

            val_progress = tqdm(val_dataloader, desc="Validating", leave=False)

            with torch.no_grad():
                for context_idx, target_idx, negative_idx in val_progress:
                    context_idx = context_idx.to(device)
                    target_idx = target_idx.to(device)
                    negative_idx = negative_idx.to(device)

                    loss = self(context_idx, target_idx, negative_idx)
                    total_val_loss += loss.item()

                    val_progress.set_postfix(val_loss=loss.item())

            avg_val_loss = total_val_loss / len(val_dataloader)
            val_losses.append(avg_val_loss)

            print(f"Train Loss: {avg_train_loss:.4f} | Validation Loss: {avg_val_loss:.4f}")

        return train_losses, val_losses

    def evaluate_test(self, test_dataloader, device='cpu'):
        self.eval()
        total_test_loss = 0
        with torch.no_grad():
            for context_idx, target_idx, negative_idx in test_dataloader:
                context_idx, target_idx, negative_idx = context_idx.to(device), target_idx.to(device), negative_idx.to(device)
                loss = self(context_idx, target_idx, negative_idx)
                total_test_loss += loss.item()

        avg_test_loss = total_test_loss / len(test_dataloader)
        return avg_test_loss

    def get_word_embeddings(self):
        return self.embedding.weight.data

    def get_target_embeddings(self):
        return self.context_layer.weight.data
```

```python
import torch
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

def get_top_similar_words(model, word, word2idx, idx2word, top_n=5):
    word_idx = word2idx[word]
    
    # Convert embeddings to numpy using .tolist() workaround
    embedding_vector = np.array(model.get_word_embeddings()[word_idx].cpu().detach().tolist())
    all_embeddings = np.array(model.get_word_embeddings().cpu().detach().tolist())
    
    similarities = cosine_similarity(
        [embedding_vector],
        all_embeddings
    )[0]
    
    top_indices = similarities.argsort()[-(top_n + 1):-1][::-1]
    
    top_words = [idx2word[i] for i in top_indices]
    
    return top_words

print("CBOW model")
for word in words_chosen:
    top_words = get_top_similar_words(cbow_model, word, word2idx_dict, idx2word_dict, top_n=5)
    print(f"Top 5 similar words to '{word}': {top_words}")
```

```python
from torch.utils.data import Dataset, DataLoader
import torch

class NewsDataset(Dataset):
    def __init__(self, df, model=None):
        """
        Args:
            df (DataFrame): The dataframe containing the data
            model: FastText model (if you need to generate embeddings here, but we assume it's precomputed)
        """
        self.texts = df['embedding'].tolist() 
        self.labels = df['label'].tolist()    
        
    def __len__(self):
        return len(self.texts)
    
    def __getitem__(self, idx):
        text_tensor = torch.tensor(self.texts[idx], dtype=torch.float32)
        label_tensor = torch.tensor(self.labels[idx], dtype=torch.long)
        return text_tensor, label_tensor
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from tqdm import tqdm

class NewsClassifier(nn.Module):
    def __init__(self, input_dim, hidden_dims=[512, 256, 128], output_dim=4):
        super(NewsClassifier, self).__init__()
        
        self.fc1 = nn.Linear(input_dim, hidden_dims[0]) 
        self.dropout1 = nn.Dropout(0.5)
        self.fc2 = nn.Linear(hidden_dims[0], hidden_dims[1]) 
        self.dropout2 = nn.Dropout(0.5)  
        self.fc3 = nn.Linear(hidden_dims[1], hidden_dims[2])  
        self.dropout3 = nn.Dropout(0.5) 
        
        self.fc_out = nn.Linear(hidden_dims[2], output_dim)
        
    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = self.dropout1(x) 
        x = F.relu(self.fc2(x))
        x = self.dropout2(x) 
        x = F.relu(self.fc3(x))
        x = self.dropout3(x) 
        
        x = self.fc_out(x)  
        return F.log_softmax(x, dim=1) 
```

```python
model = NewsClassifier(input_dim=INPUT_DIM)

criterion = nn.CrossEntropyLoss() 
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE, weight_decay=WEIGHT_DECAY)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
```

```python
def train(model, train_loader, criterion, optimizer, device):
    model.train()
    running_loss = 0
    correct_preds = 0
    total_preds = 0

    for batch in tqdm(train_loader, desc="Training", leave=False):
        embeddings, labels = batch
        embeddings, labels = embeddings.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(embeddings)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        _, predicted = torch.max(outputs, 1)
        correct_preds += (predicted == labels).sum().item()
        total_preds += labels.size(0)

    avg_loss = running_loss / len(train_loader)
    accuracy = correct_preds / total_preds * 100
    return avg_loss, accuracy

def evaluate(model, val_loader, criterion, device):
    model.eval()
    running_loss = 0
    correct_preds = 0
    total_preds = 0

    with torch.no_grad():
        for batch in tqdm(val_loader, desc="Evaluating", leave=False):
            embeddings, labels = batch
            embeddings, labels = embeddings.to(device), labels.to(device)

            outputs = model(embeddings)
            loss = criterion(outputs, labels)
            running_loss += loss.item()

            _, predicted = torch.max(outputs, 1)
            correct_preds += (predicted == labels).sum().item()
            total_preds += labels.size(0)

    avg_loss = running_loss / len(val_loader)
    accuracy = correct_preds / total_preds * 100
    return avg_loss, accuracy
```
